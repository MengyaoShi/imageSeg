```python
import numpy as np
def extract_bboxes(mask): # m:mask
    """Compute bounding boxes from masks.
    mask: [height, width, num_instances]. Mask pixels are either 1 or 0.
    Returns: bbox array [num_instances, (y1, x1, y2, x2)].
    """

    boxes = np.zeros([1, 4], dtype=np.int32)
    h, w = mask.shape[:2]
    # Bounding box.
    horizontal_indicies = np.where(np.any(mask, axis=0))[0]
    vertical_indicies = np.where(np.any(mask, axis=1))[0]
    if horizontal_indicies.shape[0]:
        x1, x2 = horizontal_indicies[[0, -1]]
        y1, y2 = vertical_indicies[[0, -1]]
        # x2 and y2 should not be part of the box. Increment by 1.
        x2 += 1
        y2 += 1
    else:
        # No mask for this instance. Might happen due to
        # resizing or cropping. Set bbox to zeros
        x1, x2, y1, y2 = 0, 0, 0, 0
    x1, y1 = max(x1-4,0), max(y1-4,0) # add a margin 
    x2, y2 = min(x2+4,w), min(y2+4,h)

    boxes[0] = np.array([y1, x1, y2, x2])
    return boxes.astype(np.int32)
```

```python
train_path = "../data/stage1_train"
img_ids = os.listdir(train_path)
```

```python
mask_paths = glob.glob(os.path.join(train_path, img_id,"masks", "*.png"))
```

```python
coordinates = []
patch_masks = []

for i in range(len(mask_paths)):
    img_id = mask_paths[i].split("/")[-3]
    img_path = glob.glob(os.path.join(train_path,img_id,"images","**"))[0]
    img_arr = cv2.imread(img_path)
    mask = cv2.imread(mask_paths[i])
    mask = (mask!=0)*1.0
    bbox = extract_bboxes(mask[:,:,0])[0]
    x1,y1,x2,y2 = bbox
    patch = img_arr[x1:x2,y1:y2]
    patch_mask = mask[x1:x2,y1:y2]
    coordinates.append(bbox)
    patch_masks.append(patch_mask)

coor_masks = {"coor":coordinates,"mask":patch_masks}
```

```python
import pickle
pickle.dump( coor_masks, open( "../sample_mask/x1y1_x2y2_masks.p", "wb" ) )
```
