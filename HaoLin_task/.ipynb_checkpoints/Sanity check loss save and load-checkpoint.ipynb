```python
import numpy as np
from PIL import Image, ImageDraw

def duplicateDim(mask):
    h, w = mask.shape
    mask3 = np.ones((h,w,3))
    mask3[:,:,0] = mask
    mask3[:,:,1] = mask
    mask3[:,:,2] = mask
    return mask3

def stackImage(file_path, **kwargs):
    #kwargs is a dict holding  mask,img,label,file_path
    num = len(kwargs)
    img = kwargs['img']
    k = 4 #k is a margin for separating image from mask
    
    frame = np.full((img.shape[0],k,3),255)
    frame[:,:,0] =0

    if num == 3:
       stack = np.hstack((kwargs['mask']*255,frame, kwargs['img']*255, frame, kwargs['label']*255))
    else:
       stack = np.hstack((kwargs['mask'],frame, kwargs['img']))

    img_back = Image.fromarray(np.uint8(stack))
    
    d = ImageDraw.Draw(img_back)
    d.text((10,10), "Prediction", fill=(255,255,0))
    d.text((10+img.shape[1]+k,10), "Image", fill=(255,255,0))
    d.text((10+img.shape[1]*2+k*2,10), "Mask", fill=(255,255,0))
    
    img_back.save(file_path)
```

```python
from scipy.stats import logistic

class Train(object):
    def __init__(self,model,criterion):
        self.model = model
        self.criterion = criterion
    def one_epoch(self,trainloader, lr = 0.0001):
        optimizer = optim.Adam(net.parameters(), lr=lr)
        for i, data in enumerate(trainloader, 0):
            inputs, labels = data['image'],data['mask']
            inputs, labels = Variable(inputs.cuda()), Variable(labels.cuda())
            optimizer.zero_grad()

            outputs = self.model(inputs)
            loss = self.criterion(outputs, labels)
            loss.backward()
            optimizer.step()
        #print("training loss: ", loss.data[0])
    def fit(self,trainloader,valloader=None,lr=0.001,epochs=3):
        for epoch in range(epochs):
            self.one_epoch(trainloader,lr)
            self.evaluate(valloader)
            
    def predict_batch(self,inputs):
        inputs = Variable(inputs,volatile = True).cuda()
        preds = self.model(inputs)
        preds = preds.cpu()
        preds = preds.data.numpy()
        preds = logistic.cdf(preds)
        return preds
    
    def predict(self,testloader):
        masks = []
        for i, data in enumerate(testloader, 0):
             masks.append(self.predict_batch(data))
        return masks
        
    def evaluate(self, valloader,probe=False):
        losses = []
        masks = []
        for i, data in enumerate(valloader, 0):
            inputs, labels = data['image'],data['mask']
            inputs, labels = Variable(inputs,volatile = True).cuda(), Variable(labels, volatile=True).cuda()
            outputs = self.model(inputs)
            
            preds = outputs.cpu()
            preds = preds.data.numpy()
            preds = logistic.cdf(preds)
            masks.append(preds)
            
            loss = self.criterion(outputs, labels).data[0]
            losses.append(loss)
        print("val loss", sum(losses)/len(losses))
        return masks if probe else losses
    
    def visualizePred(self,valloader,out_pred_path): # to merge this function with the fun on test
        for i, data in enumerate(valloader, 0):
            imgs, labels = data['image'],data['mask']
            inputs = Variable(imgs,volatile = True).cuda()

            preds = self.model(inputs.float())
            preds = preds.cpu()
            preds = preds.data.numpy()
            preds = logistic.cdf(preds)

            mask = preds> 0.5
            mask = mask[0,0,:,:]*1 


            img = imgs.numpy()[0]
            img = img.transpose((1,2,0))


            label = labels.numpy()[0]
            label3 = duplicateDim(label)

            mask3 = duplicateDim(mask)

            file_path = out_pred_path +"/"+str(i)+".jpg"

            stackImage(file_path, mask = mask3,img = img,label = label3)
            
            
    def visualizeTestPred(self,testloaderr,out_pred_path): # to merge this function with the fun on test
        for i, imgs in enumerate(testloader, 0):
            preds = self.predict_batch(imgs)

            mask = preds> 0.5
            mask = mask[0,0,:,:]*1 
            mask3 = duplicateDim(mask)
            mask3 = mask3*255

            img = imgs.numpy()[0]
            img = img.transpose((1,2,0))*255


            file_path = out_pred_path +"/"+str(i)+".jpg"

            stackImage(file_path,mask=mask3,img=img)
```

```python
def rle_encoding(x):
    dots = np.where(x.T.flatten() == 1)[0]
    run_lengths = []
    prev = -2
    for b in dots:
        if (b>prev+1): run_lengths.extend((b + 1, 0))
        run_lengths[-1] += 1
        prev = b
    return run_lengths

def label_to_rles(lab_img):
    for i in range(1, lab_img.max() + 1):
        yield rle_encoding(lab_img == i)
```

```python
import pandas as pd
from skimage.morphology import label

def predictTest(net,testloader):
    masks = []
    for i, data in enumerate(testloader, 0):
    
        inputs = Variable(data,volatile = True).cuda()
        outputs = net(inputs) 
        outputs = net(inputs)
        preds = outputs.cpu()
        arrs = preds.data.numpy()
        mask = logistic.cdf(arrs)
        mask = (mask > 0.5)*1
        masks.append(mask)
    return masks

def mask2RLE(masks,test_path):
 
    test_imgs = os.listdir(test_path)
    
    rles = []
    new_test_ids = []
    for mask,img_name in zip(masks,test_imgs):

        img = cv2.imread(test_path+"/"+img_name)
        size = img.shape[0:2]
        mask = np.squeeze(mask)

        mask = resize(mask,size,mode='constant', preserve_range=True)
        lab_img = label(mask > 0)
        rle = list(label_to_rles(lab_img))
        rles.extend(rle)
        new_test_ids.extend([img_name[:-4]] * len(rle))
    return (rles,new_test_ids)

def makesub(rles, new_test_ids,file_path):
    sub = pd.DataFrame()
    sub['ImageId'] = new_test_ids
    sub['EncodedPixels'] = pd.Series(rles).apply(lambda x: ' '.join(str(y) for y in x))
    sub.to_csv(file_path, index=False)
    print(sub.shape)
    print("submission file generated in ",file_path)
    
test_path = root+"/test"
file_path = sub_path+'/LinkCropLoss0.904.csv'

masks = predictTest(train.model,testloader)
rles,new_test_ids = mask2RLE(masks,test_path)
makesub(rles, new_test_ids,file_path)
```
