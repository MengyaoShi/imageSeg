```python
import os
import cv2
from utils.probeImage import probeImage
from PIL import Image,  ImageDraw
import numpy as np
import pickle
import glob
```

```python
# to do create a module for generate a mask for each object
'''
Inputs: 
1. An image: [h, w, 3]; 
2. coor_masks: a dict of two keys for the image. 
   the value "coor" holds the coordiantes of the bounding box
   the corresponding segmentation is stored the other value named "mask"
   
Outputs:
An array: [h,w,num]; num is equal to the length of coor_masks['coor'] (the number of objects in this image)


Note:
If two nuclei overlap, the overlapping pixels can only belong to one mask. Otherwise, the submission 
an Exception will pop out when submitting the solution. 
'''
```

```python
img_path = glob.glob(os.path.join("../sample_mask","**png"))[0]
coor_masks = pickle.load( open( "../sample_mask/x1y1_x2y2_masks.p", "rb" ) )
xys = coor_masks["coor"]
masks = coor_masks["mask"]
```

```python
img = Image.open(img_path).convert("RGB")
img_arr = np.asarray(img, order='F')
```

```python
# the 4th mask overlaps with other objects
i = 3
bbox = xys[i]
x1, y1, x2, y2 = bbox
mask = masks[i]
```

```python
#visualize an object
probeImage(img_arr[x1:x2,y1:y2])
```

```python
probeImage(mask)
```

```python

draw = ImageDraw.Draw(img)
# the coordinates used in PILLOW seems y1,x1, y2,x2
draw.rectangle(((bbox[1],bbox[0]),(bbox[3],bbox[2])), fill=None, outline=(255, 0, 0))
img
```
